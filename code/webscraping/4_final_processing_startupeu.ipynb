```python
# 1) Define the columns you want to keep
keep_cols = [
    "name_cleaned",
    "link_startupeu",
    "link_logo",
    "website_host",
    "website_up",
    "city",
    "region",
    "category",
    "top_3_keywords",
    "business_model",
    "linkedin_global_profile",
    "linkedin_handle",
    "valid_linkedin",
    "instagram_handle",
    "valid_instagram",
    "x_handle",
    "valid_x",
    "founded",
    "age",
    "total_funding",
    "n_valid_headlines",
    "valid_headlines",
    "success_count",
    "total_funding_eur",
    "acquisition_count",
    "bankruptcy_count",
    "non_success_count",
    "writing_score",
    "clarity_score",
    "innovativeness_score",
    "market_readiness_score",
    "founder_signal_score",
    "sentiment_score",
    "traction_score",
    "word_count",
    "sentence_count",
    "jargon_density",
    "numeric_evidence_count",
    "call_to_action_flag"
]

# 2) Subset your DataFrame
df_small = df_listings[keep_cols].copy()

# 3) Define the rename mapping
rename_map = {
    "name_cleaned": "name",
    "website_host": "link_website",
    "valid_linkedin": "linkedin_valid",
    "valid_instagram": "instagram_valid",
    "valid_x": "x_valid",
    "founded": "founded_year",
    "total_funding": "total_funding_listing",
    "n_valid_headlines": "headline_count",
    "valid_headlines": "headlines",
    "success_count": "headline_financial_count",
    "total_funding_eur": "headline_funding",
    "acquisition_count": "headline_acquisition_count",
    "bankruptcy_count": "headline_bankruptcy_count",
    "non_success_count": "headline_non_financial_count"
}

# 4) Apply the renaming
df_small.rename(columns=rename_map, inplace=True)

# 5) Inspect the result
print(df_small.columns.tolist())
df_small.head()
```

**Remove Invalid Handles**

```python
# For any handle where the corresponding “_valid” flag is False, set that handle to NaN
df_small.loc[~df_small["linkedin_valid"], "linkedin_handle"]   = pd.NA
df_small.loc[~df_small["instagram_valid"], "instagram_handle"] = pd.NA
df_small.loc[~df_small["x_valid"], "x_handle"]                 = pd.NA

# (Optional) inspect a few rows to confirm
print(df_small[[
    "name",
    "linkedin_handle", "linkedin_valid",
    "instagram_handle", "instagram_valid",
    "x_handle",       "x_valid"
]].head())
```

**Remove down websites**

```python
# Set link_website to NaN wherever website_up is False
df_small.loc[~df_small["website_up"], "link_website"] = pd.NA

# Optional: verify the change
print(df_small[["name", "link_website", "website_up"]].head())
```

**Remove duplicates**

```python
import pandas as pd

# 1) Build a mask of rows to drop: any duplicate non-null handle in x, instagram, or linkedin
mask_dup = pd.Series(False, index=df_small.index)
for col in ["x_handle", "instagram_handle", "linkedin_handle"]:
    dup = df_small[col].duplicated(keep="first") & df_small[col].notna()
    mask_dup |= dup

# 2) Drop those rows, keep only the first occurrence per handle
df_small = df_small.loc[~mask_dup].reset_index(drop=True)

# 3) (Optional) Inspect how many were removed
print(f"Dropped {mask_dup.sum()} rows with duplicate handles.")

# 4) Verify no duplicates remain
for col in ["x_handle", "instagram_handle", "linkedin_handle"]:
    print(f"Remaining unique {col}: {df_small[col].nunique(dropna=True)}")
```

```python
import pandas as pd

# 1) Identify duplicate link_website values (excluding NaN), keeping only the first occurrence
dup_web = df_small["link_website"].duplicated(keep="first") & df_small["link_website"].notna()

# 2) Drop those duplicate rows
df_small = df_small.loc[~dup_web].reset_index(drop=True)

# 3) (Optional) Inspect how many were removed
print(f"Dropped {dup_web.sum()} rows with duplicate link_website.")

# 4) Verify no duplicates remain
print(f"Remaining unique link_website: {df_small['link_website'].nunique(dropna=True)}")
```

**Check Top 10 values**

```python
top10_data = {}

for col in df_small.columns:
    # include NaNs, then take the top 10
    vc = df_small[col] \
         .value_counts(dropna=False) \
         .head(10)

    # (optional) relabel the NaN index to "<Missing>"
    vc.index = vc.index.where(~vc.index.isna(), '<Missing>')

    top_values = vc.index.tolist()
    top_counts = vc.values.tolist()

    top10_data[col]       = pd.Series(top_values)
    top10_data[f"{col}_count"] = pd.Series(top_counts)

top10_df = pd.DataFrame(top10_data)

display(top10_df)
```

## Create Success Variable

### Closed Startups

```python
# Define the columns you want to inspect
cols = [
    "name",
    "linkedin_valid",
    "instagram_valid",
    "x_valid",
    "total_funding_listing",
    "headline_count",
    "headline_financial_count",
    "headline_funding",
    "headline_acquisition_count",
    "headline_bankruptcy_count"
]

# Subset the bad_websites DataFrame
bad_websites = df_small[df_small["website_up"] == False]

# Display the selected columns for inspection
print(f"Entries with website_up = False: {len(bad_websites)}\n")
display(bad_websites[cols])
```

**Create Status column and adjust status based on website_up = False:**

website down = inactive   
website down + bankrupcy headline = bankrupt

```python
import numpy as np

# 1) Initialize the new column with NaN
df_small["status"] = np.nan

# 2) Build masks
mask_down    = df_small["website_up"] == False
mask_bankrupt = mask_down & (df_small["headline_bankruptcy_count"] >= 1)

# 3) Assign “bankrupt” where site is down AND at least one bankruptcy headline
df_small.loc[mask_bankrupt, "status"] = "bankrupt"

# 4) Assign “inactive” where site is down but no bankruptcy headlines
df_small.loc[mask_down & ~mask_bankrupt, "status"] = "inactive"

# 5) (Optional) Verify
print(df_small[["name","website_up","headline_bankruptcy_count","status"]].head(10))
```
