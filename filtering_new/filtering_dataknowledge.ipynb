```python
import pandas as pd 
```

```python
startup_df = pd.read_csv("../data_new/startups_final.csv")
startup_df
```

```python
cols_to_keep = ["name",
                "city",
                "region",
                "category",
                "business_model",
                "status"
                ]

df_startup_1 = startup_df[cols_to_keep]

df_startup_1.head()
```

```python
# Filter out inactive startups 

df_startup_2 = df_startup_1.query("status != 'inactive'")
```

```python
df_startup_2
```

```python
vienna = df_startup_2[df_startup_2['region'] == 'Vienna'].sample(n=50, random_state=1)
non_vienna = df_startup_2[df_startup_2['region'] != 'Vienna'].sample(n=100, random_state=1)

df_startup_3 = pd.concat([vienna, non_vienna])
```

```python
pd.set_option('display.max_rows', None)
```

```python
startups_remove = [
    "Enpulsion",
    "Hyrox Training Plans",
    "RebeccAi",
    "Lignovations",
    "ForgTin",
    "sproof",
    "Fermify",
    "CoLivi",
    "boomerank",
    "Ohana",
    "Wood Space",
    "myAcker",
    "Easy Month End"
]

df_startup_4 = df_startup_3[~df_startup_3['name'].isin(startups_remove)]


df_startup_4.reset_index(drop=True, inplace=True)
```

```python
df_startup_4

df_startup_4.to_csv("./data/startups_data_and_knowledge.csv", index=False)
```
