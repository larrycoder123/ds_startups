```python
# Filter out inactive startups 

startup_df_1 = startup_df.query("status != 'inactive'")
```

```python
startup_df_3 = startup_df_2.query("status == 'funding'")

startup_df_3
```

```python
regions_keep = ["Vienna"]

startup_df_4 = startup_df_3[startup_df_3["region"].isin(regions_keep)]
```
